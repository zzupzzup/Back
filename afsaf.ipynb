```python
import pandas as pd
def input_log(stores):    
    # 가게 이름 입력 받기
    store_name = input("가게 이름을 입력하세요 (여러 개의 가게 이름은 쉼표로 구분): ")
    store_name_list = store_name.split(",")
    
    # 검색 결과를 담을 빈 데이터프레임 생성
    results = pd.DataFrame(columns=['store', 'point', 'view', 'review', 'star', 'address', 'category', 'join_key', 'menu', 'x', 'y'])

    # 가게 정보 검색
    for name in store_name_list:
        result = stores[stores['store'] == name.strip()]  # 가게 이름 앞뒤 공백 제거
        results = pd.concat([results, result])

    # 검색 결과 출력
    return results
```
